```python
import requests

URL = "http://andrewbeatty1.pythonanywhere.com/books"

def get_books():
    response = requests.get(URL)
    if response.status_code == 200:
        return response.json()
    else:
        print("Error fetching books")
        return []

def calculate_average_price(books):
    total = 0
    count = 0

    for book in books:
        if book["price"] is not None:
            total += float(book["price"])
            count += 1

    if count == 0:
        return 0

    return total / count


if __name__ == "__main__":
    books = get_books()
    average = calculate_average_price(books)
    print(f"Average book price: {average:.2f}")
```
