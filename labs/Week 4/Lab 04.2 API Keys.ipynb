```python
import requests
import json
from Config import config as cfg

apiKey = cfg["github"]

headers = {
    "Authorization": f"Bearer {apiKey}",
    "Accept": "application/vnd.github+json"
}

owner = "AudeAllen"
repo = "APrivateOne"

#  Get repository metadata
repo_url = f"https://api.github.com/repos/{owner}/{repo}"
repo_response = requests.get(repo_url, headers=headers)
repo_data = repo_response.json()

#  Get default branch
default_branch = repo_data["default_branch"]

tree_url = f"https://api.github.com/repos/{owner}/{repo}/git/trees/{default_branch}?recursive=1"
tree_response = requests.get(tree_url, headers=headers)
tree_data = tree_response.json()

# Combine everything
combined_data = {
    "repository_info": repo_data,
    "file_tree": tree_data
}

#  Write to file
filename = "repo_output.json"

with open(filename, "w") as fp:
    json.dump(combined_data, fp, indent=4)

print("Saved successfully.")
```

### Make a change to the repository in Github


```python
import requests
import base64
import json
from Config import config as cfg

# GitHub token
apiKey = cfg["github"]

headers = {
    "Authorization": f"Bearer {apiKey}",
    "Accept": "application/vnd.github+json"
}

owner = "AudeAllen"
repo = "APrivateOne"
file_path = "api_test.txt"

# File content
content = "This file was created using the GitHub API.\n"

# Encode content to base64 (required by GitHub)
encoded_content = base64.b64encode(content.encode("utf-8")).decode("utf-8")

# API endpoint
url = f"https://api.github.com/repos/{owner}/{repo}/contents/{file_path}"

# Data payload
data = {
    "message": "Creating api_test.txt via API",
    "content": encoded_content
}

# Send PUT request (creates file)
response = requests.put(url, headers=headers, data=json.dumps(data))

print(response.status_code)
print(response.json())
```
