```python
import requests
import csv
from xml.dom.minidom import parseString

# Download XML data
url = "http://api.irishrail.ie/realtime/realtime.asmx/getCurrentTrainsXML"
response = requests.get(url)
doc = parseString(response.content)

# Tags to retrieve
retrieveTags = [
    'TrainStatus',
    'TrainLatitude',
    'TrainLongitude',
    'TrainCode',
    'TrainDate',
    'PublicMessage',
    'Direction'
]


objTrainPositionsNodes = doc.getElementsByTagName("objTrainPositions")

# Open CSV file and write data
with open('week03_train.csv', mode='w', newline='') as train_file:
    train_writer = csv.writer(train_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)

    # Write the header row
    train_writer.writerow(retrieveTags)

    # Loop through each train
    for objTrainPositionsNode in objTrainPositionsNodes:
      
        trainCodeNode = objTrainPositionsNode.getElementsByTagName("TrainCode").item(0)
        if trainCodeNode is None or trainCodeNode.firstChild is None:
            continue
        trainCode = trainCodeNode.firstChild.nodeValue.strip()
     

     
        dataList = []
        for retrieveTag in retrieveTags:
            datanode = objTrainPositionsNode.getElementsByTagName(retrieveTag).item(0)
            if datanode is not None and datanode.firstChild is not None:
                dataList.append(datanode.firstChild.nodeValue.strip())
            else:
                dataList.append('') 

        # Write to CSV
        train_writer.writerow(dataList)
```

### Only show Trains beginning with a 'D' Code 

```python
import requests
import csv
from xml.dom.minidom import parseString

# Download XML data
url = "http://api.irishrail.ie/realtime/realtime.asmx/getCurrentTrainsXML"
response = requests.get(url)
doc = parseString(response.content)

# Tags to retrieve
retrieveTags = [
    'TrainStatus',
    'TrainLatitude',
    'TrainLongitude',
    'TrainCode',
    'TrainDate',
    'PublicMessage',
    'Direction'
]


objTrainPositionsNodes = doc.getElementsByTagName("objTrainPositions")

# Open CSV file and write data
with open('week03_train_train_D.csv', mode='w', newline='') as train_file:
    train_writer = csv.writer(train_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)

    # Write the header row
    train_writer.writerow(retrieveTags)

    # Loop through each train
    for objTrainPositionsNode in objTrainPositionsNodes:
      
        trainCodeNode = objTrainPositionsNode.getElementsByTagName("TrainCode").item(0)
        if trainCodeNode is None or trainCodeNode.firstChild is None:
            continue
        trainCode = trainCodeNode.firstChild.nodeValue.strip()
        if not trainCode.startswith('D'):
            continue

       
     
        dataList = []
        for retrieveTag in retrieveTags:
            datanode = objTrainPositionsNode.getElementsByTagName(retrieveTag).item(0)
            if datanode is not None and datanode.firstChild is not None:
                dataList.append(datanode.firstChild.nodeValue.strip())
            else:
                dataList.append('') 

        # Write to CSV
        train_writer.writerow(dataList)
```
