```python
import requests
import urllib3
from collections import Counter

#Want to suppress the warning for SSL - 

urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

# https://github.com/CiscoDevNet/dne-dna-code/blob/master/rest-api/python/deck_of_cards.py
# https://stackoverflow.com/questions/63542770/print-multiple-entries-from-a-json-response-to-e-get-request

# Shuffle a new deck
url1 = "https://deckofcardsapi.com/api/deck/new/shuffle/?deck_count=1"
response = requests.get(url1, verify=False)
data = response.json()
deck_id = data['deck_id']

# Draw 5 cards from the deck
url2 = f"https://deckofcardsapi.com/api/deck/{deck_id}/draw/?count=5"
response = requests.get(url2,verify=False)
cards_data = response.json()
cards = cards_data['cards']

# Print out each card
print("You drew the following cards:\n")
values = []
suits = []

for card in cards:
    value = card['value']
    suit = card['suit']
    print(f"{value} of {suit}")
    values.append(value)
    suits.append(suit)


# Check for pair, triple, straight, or flush

# Pair = 2 cards of same value
# Triple = 3 cards of same value
# Straight = 5 cards in sequence
# Flush = all same suit

# Convert face cards to numbers as this will be easier to match - ACE, King etc.

value_map = {
    "ACE": 14, "KING": 13, "QUEEN": 12, "JACK": 11,
    "10": 10, "9": 9, "8": 8, "7": 7, "6": 6,
    "5": 5, "4": 4, "3": 3, "2": 2
}
numbers = [value_map[v] for v in values]
numbers.sort()

# Check for duplicates
counts = Counter(values)
has_pair = 2 in counts.values()
has_triple = 3 in counts.values()

# Check for flush
flush = all(s == suits[0] for s in suits)

# Check for straight
straight = all(numbers[i] + 1 == numbers[i+1] for i in range(4))

# Print result and congratulate the user
print("\nResult:")
if flush:
    print("Congratulations! You have a flush!")
if straight:
    print("Congratulations! You have a straight!")
if has_triple:
    print("Congratulations! You have three of a kind!")
elif has_pair:
    print("Congratulations! You have a pair!")
else:
    print("Nothing this time. Try again!")
```
